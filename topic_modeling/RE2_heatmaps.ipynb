#Comparisons between *Resident Evil 2* (1998) and *Resident Evil 2 Remake* (2019)
Although this project is exploratory, there are several observations about the roles of the female characters based on the heat maps. Focusing on Claire in the original *Resident Evil 2*, the heat map places Topic 3 as the most relevant to her dialogue. Topic 3 contained words such as "sherry," "leon," "chris," "annette," "save," and "brother." ChatGPT interpreted this topic as relating to familial relationships and finding someone. Claire's role in the game pertains most to finding her brother and becoming a surrogate mother to Sherry. In *Resident Evil 2 Remake*, Topic 7 is the most emphasized, followed by Topic 4. ChatGPT interpreted Topic 7 as focusing on Sherry and moments of tension, highlighted by emotional reactions and explicit language. Topic 4 was less relevant to Claire, and ChatGPT's interpretation focused on interactions with Leon with moments of hesitation and realization. This topic also contained keywords like "brother." Although Claire's role in the game remains Sherry's savior, her ties to male characters are less stressed than in the previous release.

Leon's relationship to Ada seems to change between both games according to the heatmaps. In *Resident Evil 2*, "ada" is the first keyword in Topic 9, which is the most relevant to Leon. ChatGPT interpreted this topic to be about Ada and emphasizing moments of connection. In *Resident Evil 2 Remake*'s heatmap, none of the topics with a weight greater than .08 has "ada" as a keyword. None of the topics seem to emphasize a connections to other characters outside of Claire, Marvin, Sherry, and Sherry's parents. Comparing the strongest topics in Ada from the original to the remake, we can see this change again. In the first game, Topic 6, which emphasizes Leon's role, is the most relevant to Ada. In the remake, Topic 8, which relates to the G-Virus and the conspiracy surrounding it, is most relevant to Ada although Topic 4, which emphasizes interactions with Leon, is the second-most relevant. This data could suggest that the writers gave Ada's character more depth with more emphasis placed on her own goals as her role moved away from propping up the importance of Leon's.
