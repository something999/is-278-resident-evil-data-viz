#Word Clouds

Leon and Claire from *Resident Evil 2* and *Resident Evil 2 Remake* were rendered into word clouds. First, full body shots of each character were downloaded from a *Resident Evil* fandom website, and then the background of the images was removed on Remove.ig. The background was removed to improve word placement within the character's silhouette. Next, the pictures were uploaded onto Wordclouds.com, and that website's interface was used to create the word clouds. The characters' clothes are the basis of the word colors, and *Resident Evil*'s font styles influenced the word clouds'. MALLET generated the words used, and the NLP determined each keyword's weight (relevance) to the character. The weights were decimal numbers smaller than one but higher than zero, but Wordclouds.com only recognizes whole numbers. Linearly transformations changed these decimals into whole numbers ranging from one to ten. Spreadsheets were created for each character with weights in the left column and the corresponding word in the right column according to a format that Wordclouds can understand.

Below I've placed the word clouds to the right of the original image in the following order: Claire (RE2), Leon (RE2), Claire (RE2 Remake), and Leon (RE2 Remake).

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load your images (replace with your actual file paths)
image_files = ['Claire_RE2.png', 'Claire_RE2_wc.png', 'Leon_RE2.png', 'Leon_RE2_wc.png',
               'Claire_RE2remake.png', 'Claire_RE2remake_wc.png', 'Leon_RE2remake.png', 'Leon_RE2remake_wc.png']
images = [mpimg.imread(file) for file in image_files]

# Create a figure and axes
fig, axes = plt.subplots(4, 2, figsize=(10, 20))  # 4 rows, 2 columns

# Display images on the axes
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.axis('off')

plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```
